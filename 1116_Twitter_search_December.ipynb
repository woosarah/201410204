## T3. Twitter에 'Seoul'을 12월 한 달 분량을 읽기

```python
import os
keyPath=os.path.join(os.getcwd(),'src','twitter.properties')

def getKey(keyPath):
    d=dict()
    f=open(keyPath,'r')
    lines=f.readlines()
    for line in lines:
        row=line.split('=')
        row0=row[0]
        d[row0]=row[1].strip()
    return d

key=getKey(keyPath)
```

```python
import oauth2 as oauth
import json

consumer = oauth.Consumer(key=key['CONSUMERKEY'], secret=key['CONSUMERSECRET'])
token=oauth.Token(key=key['ACCESSTOKEN'], secret=key['ACCESSTOKENSECRET'])
```

```python
client = oauth.Client(consumer, token)
```
